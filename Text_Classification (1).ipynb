```python
import pandas as pd
import numpy as np
import csv
```

```python
temp_df = pd.read_csv('/content/IMDB Dataset (1).csv', on_bad_lines='skip', quoting=csv.QUOTE_MINIMAL)
```

```python
df = temp_df.iloc[:10000]
```

```python
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Basic Preprocessing
# Remove Tags
# Lowecase
# remove stopwords
import re
def remove_tags(raw_text):
    cleaned_text = re.sub(re.compile('<.*?>'), '', raw_text)
    return cleaned_text
```

```python
df['review'] = df['review'].apply(remove_tags)
```

```python
df['review'] = df['review'].apply(lambda x:x.lower())
```

```python
df
```

```python
df['review'][1]
```

```python
# stop words
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

sw_list = stopwords.words('english')

df['review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

```python
X = df.iloc[:,0:1]
y = df['sentiment']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_train_bow.shape
```

```python
X_test_bow.shape
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
```

```python
gnb.fit(X_train_bow, y_train)
```

```python
y_pred = gnb.predict(X_test_bow)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

Using Tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(stop_words='english', max_features=5000)
```

```python
X_train_tf = tf.fit_transform(X_train['review']).toarray()
X_test_tf = tf.transform(X_test['review']).toarray()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train_tf, y_train)
```

```python
y_pred = model.predict(X_test_tf)
```

```python
from sklearn.metrics import classification_report
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
