```python
df['Sentiment_Subjectivity'] = df['Sentiment_Subjectivity'].fillna(df['Sentiment_Subjectivity'].mean())
```

```python
df['Sentiment_Polarity'] = df['Sentiment_Polarity'].fillna(df['Sentiment_Polarity'].mean())
```

```python
df_mode = df.pivot_table(values='Translated_Review', columns='Sentiment', aggfunc=(lambda x: x.mode()[0]))
df_mode
```

```python
missing_values = df['Translated_Review'].isnull()

df.loc[missing_values, 'Translated_Review'] = df.loc[missing_values, 'Sentiment'].apply(lambda x: df_mode.get(x, ''))
```

```python
missing_values = df['Sentiment'].isnull()

df.loc[missing_values, 'Sentiment'] = df.loc[missing_values, 'Translated_Review'].apply(lambda x: df_mode.get(x, ''))
```
